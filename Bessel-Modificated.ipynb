# Creación de la Gráfica de las funciones de Bessel Modificadas, para el caso homogeneo
La solución para el caso homogeneo, se tiene 
$$U(r) = AI_{n + 1/2}(\lambda r) + BK_{n+1/2}(\lambda r)$$


```python
%matplotlib inline

import numpy
import scipy.special 
import matplotlib.pyplot as plt

x = numpy.linspace(0.01, 4, 25)

for v in numpy.linspace(0,3,4):
    plt.plot(x, scipy.special.iv(v,x))
#plt.plot(x, scipy.special.yv(v,x))
plt.grid()
plt.show()

```

```python
x = numpy.linspace(0.005, 4, 25)

for v in numpy.linspace(0,2,1):
    plt.plot(x, scipy.special.kv(v,x), label = v )
#plt.plot(x, scipy.special.yv(v,x))
plt.legend()
plt.grid()
plt.show()
```

```python
x = numpy.linspace(0.1, 4, 25)

for v in numpy.linspace(0,3,4):
    plt.plot(x, scipy.special.iv(v,x)/numpy.sqrt(x), label = v)
#plt.plot(x, scipy.special.yv(v,x))
plt.legend()
plt.grid()
plt.show()
```

```python
x = numpy.linspace(0.005, 0.2, 25)

for v in numpy.linspace(0,2,1):
    plt.plot(x, scipy.special.kv(v,x)/numpy.sqrt(x), label = v )
#plt.plot(x, scipy.special.yv(v,x))
plt.legend()
plt.grid()
plt.show()
```

```python
x = numpy.linspace(0.01, 6, 250)
v = 0
hambda = 1.0
solucion = scipy.special.iv(v,hambda*x)/numpy.sqrt(x) + scipy.special.kv(v,hambda*x)/numpy.sqrt(x)

plt.plot(x, solucion)
plt.grid()
plt.show()
```

```python
x = numpy.linspace(0.01, 6, 250)
v = 0
for hambda in numpy.linspace(0.6,1.2, 4):
    solucion = scipy.special.iv(v,hambda*x)/numpy.sqrt(x) + scipy.special.kv(v,hambda*x)/numpy.sqrt(x)
    plt.plot(x, solucion, label = hambda)
plt.grid()
plt.legend()
plt.show()
```
