# Estudio de la densidad puntual con el Potencial de Yukawa y Newtoniano

En este codigo se pretende mostrar las gráficas asociadas a una carga puntual, con la función de Green Newtoniana,
$$G_{New}(\vec{r}, \vec{r}')= \frac{1}{|\vec{r} - \vec{r}'|} \rightarrow \Phi_{New}(\vec{r}) = \frac{-GM}{|\vec{r}|}$$
Para el caso del potencial tipo Yukawa se tiene los sieguiente:
$$G_{Yuka}(\vec{r}, \vec{r}')= \frac{e^{\lambda|\vec{r} - \vec{r}'|}}{|\vec{r}-\vec{r}'|} \rightarrow \Phi_{Yuka}(\vec{r}) = -GM\frac{e^{\lambda|\vec{r}|}}{|\vec{r}|}$$

```python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt

def Newtoniano(r):
    # En este caso se tomo que la M del objeto es M = 1, y G = 1 con ello tenemos
    #
    Phi = 1/numpy.abs(r)
    return Phi

def Yukawa(r, Lambda):
    # En este caso se tomo que la M del objeto es M = 1, y G = 1 con ello tenemos
    #
    Phi = numpy.exp(numpy.abs(r)/Lambda)/numpy.abs(r)
    return Phi

distancia = numpy.linspace(0.001,0.03,50)

PhiNewtoniano = Newtoniano(distancia)
PhiYukawa     = Yukawa(distancia,1)

#print(PhiNewtoniano, PhiYukawa)
plt.plot(distancia, PhiNewtoniano, 'bo', label = 'Newtoniano')
plt.plot(distancia, PhiYukawa    , 'r' , label = 'Yukawa ')
plt.legend()
plt.ylabel(r'$\phi(r)$')
plt.xlabel(r'r')
plt.grid()
plt.show()
```

### Variando el parametro $\lambda$


A continuación se va a variar el parametro, $\lambda$ para determinar como afecta en el potencial 

```python
distancia = numpy.linspace(0.001,0.03,50)

for hambda in numpy.linspace(0.008,0.02,5):
    PhiYukawa = Yukawa(distancia, hambda)
    plt.plot(distancia, PhiYukawa, label= hambda)
plt.grid()
plt.ylabel(r'$\phi(r)$')
plt.xlabel(r'r')
plt.legend(loc = 'upper left')
plt.show()
```

### Cambio en la grafica de la velocidad
utilizando la ecuación para calcular la velocidad de rotación correspondiente 
$$v^{2}_{c} =r \frac{d \phi}{dr}$$
para el caso Newtoniano se tiene
$$ v_{cNew} = \sqrt{\frac{GM}{r}}$$
Para el caso del potencial tipo Yukawa se tiene
$$ v_{cYuka} = \sqrt{GM\left(\frac{e^{-r/\lambda}}{\lambda}+\frac{e^{-r/\lambda}}{r}\right)}$$


```python
def velocidadNewtoniana(r):
    v = numpy.sqrt(1/r)
    return v

def velocidadYukawa(r, hambda):
    v = numpy.sqrt((numpy.exp(-r/hambda))/hambda + (numpy.exp(-r/hambda))/r)
    return v
#print(velocidadNewtoniana(1), velocidadYukawa(1,1))
distancia = numpy.linspace(0.5, 20, 20)
hambda    = 10
plt.plot(distancia, velocidadNewtoniana(distancia)
         , label = 'Newtoniano')
plt.plot(distancia, velocidadYukawa(distancia, hambda)
         , label = 'Yukawa')
plt.legend()
plt.grid()
plt.show()
```
